```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_log_error

train_data= pd.read_csv('/content/train.csv')
test_data= pd.read_csv('/content/test.csv')

```

```python
print("Missing values in train data:\n", train_data.isnull().sum())
print("Missing values in test data:\n", test_data.isnull().sum())
```

```python
train_data.dropna(inplace=True)
test_data.dropna(inplace=True)
```

```python
X= train_data.drop(columns=['cost'])
Y= train_data['cost']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
model = LinearRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
mae = mean_absolute_error(Y_test, y_pred)
rmlse = mean_squared_log_error(Y_test, y_pred)
print("Root Mean Squared Log Error:", rmlse)
print("Mean Absolute Error:", mae)
```

```python

y_pred = model.predict(test_data)

output = pd.DataFrame({
    'id': test_data['id'],
    'cost': y_pred
})

output.to_csv('linear.csv', index=False)

print("Predictions saved to 'linear.csv'.")
```
